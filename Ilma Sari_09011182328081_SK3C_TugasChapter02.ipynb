```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Import Dataset
file_path = '/mnt/data/heart_disease_uci.csv'
df = pd.read_csv(io.StringIO(filenya['heart_disease_uci.csv'].decode('utf-8')))

# 2. Eksplorasi Awal
# Tampilkan 5 baris pertama dan terakhir dari dataset
print("5 Baris Pertama:")
print(df.head())

print("\n5 Baris Terakhir:")
print(df.tail())

# Lihat informasi tentang dataset (tipe data, jumlah entri, jumlah kolom)
print("\nInformasi Dataset:")
print(df.info())

# Cek nilai missing
print("\nCek Missing Values:")
print(df.isnull().sum())

# Cek outliers menggunakan boxplot
print("\nBoxplot untuk mendeteksi outliers:")
for col in df.select_dtypes(include=[np.number]).columns:
    sns.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

# 3. Statistik Deskriptif
# Hitung dan tampilkan mean, median, mode, standar deviasi, variansi, skewness, dan kurtosis untuk kolom numerik
desc_stats = df.describe().T
numeric_cols = df.select_dtypes(include=[np.number]).columns
desc_stats['median'] = df[numeric_cols].median()
desc_stats['mode'] = df[numeric_cols].mode().iloc[0]
desc_stats['std_dev'] = df[numeric_cols].std()
desc_stats['variance'] = df[numeric_cols].var()
desc_stats['skewness'] = df[numeric_cols].skew()
desc_stats['kurtosis'] = df[numeric_cols].kurtosis()

print("\nStatistik Deskriptif:")
print(desc_stats)

# Hitung korelasi antar variabel numerik
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
print("\nKorelasi Antar Variabel Numerik:")
print(correlation_matrix)

# 4. Visualisasi Data
# Buat histogram untuk setiap variabel numerik
print("\nHistogram untuk variabel numerik:")
for col in df.select_dtypes(include=[np.number]).columns:
    plt.hist(df[col], bins=30, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Buat box plot untuk setiap variabel numerik
print("\nBoxplot untuk variabel numerik:")
for col in df.select_dtypes(include=[np.number]).columns:
    sns.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

# Buat scatter plot untuk pasangan variabel yang memiliki korelasi tinggi (misal, > 0.5 atau < -0.5)
print("\nScatter Plot untuk pasangan variabel yang memiliki korelasi tinggi:")
high_corr_pairs = correlation_matrix[(correlation_matrix > 0.5) | (correlation_matrix < -0.5)]
high_corr_pairs = high_corr_pairs.stack().reset_index()
high_corr_pairs = high_corr_pairs[high_corr_pairs['level_0'] != high_corr_pairs['level_1']]

for index, row in high_corr_pairs.iterrows():
    plt.scatter(df[row['level_0']], df[row['level_1']])
    plt.title(f'Scatter Plot between {row["level_0"]} and {row["level_1"]}')
    plt.xlabel(row['level_0'])
    plt.ylabel(row['level_1'])
    plt.show()

# 5. Laporan
print("\nLaporan:")
print("- Dataset memiliki beberapa kolom numerik dengan distribusi yang tidak normal, ditandai dengan skewness dan kurtosis yang tinggi.")
print("- Ada beberapa outliers yang terdeteksi di beberapa kolom numerik.")
print("- Korelasi yang signifikan ditemukan antara beberapa pasangan variabel numerik, yang bisa dilihat dari scatter plot.")
print("- Tidak ada nilai missing dalam dataset.")
```
